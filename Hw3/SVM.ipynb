```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
```

```python
# 讀取train.csv和test.csv
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
train=train.drop(["Unnamed: 0"], axis=1)
test=test.drop(["Unnamed: 0"], axis=1)

# 將train和test的資料分為features和target
# train_x為features,train_y為target
# test_x為features,test_y為target
column_list=list(train.columns.values)
column_list.remove("target")
train_x=train[column_list]
train_y=train[["target"]]
test_x=test[column_list]
test_y=test[["target"]]
train_x = train_x.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
test_x = test_x.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
train_x=train_x.fillna(0)
test_x=test_x.fillna(0)

# fit model
model=SVC(kernel = 'rbf', C=10, gamma=10,probability=True)
model.fit(train_x,train_y)

# predict
predict_class=model.predict(test_x)
predict_probability=model.predict_proba(test_x)

# evaluation
accuracy=accuracy_score(test_y,predict_class)

# print result
print("Predict Class:")
print(predict_class)
print("")
print("")
print("Predict Probability:")
print(predict_probability)
print("")
print("")
print("Accuracy:")
print(accuracy)
```
