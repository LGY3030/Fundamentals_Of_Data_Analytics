```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
# 讀取data_preprocessed.csv
data=pd.read_csv("data_preprocessed.csv")
data=data.drop(["Unnamed: 0"], axis=1)

# 將資料分為train_x,train_y,test_x,test_y
columns=list(data.columns)
columns.remove('Revenue')
train_x=data[columns]
train_y=data[['Revenue']]
test_x=train_x[-1000:]
test_y=train_y[-1000:]
train_x=train_x[:-1000]
train_y=train_y[:-1000]
test_x=test_x.reset_index()
test_y=test_y.reset_index()
test_x=test_x.drop(["index"], axis=1)
test_y=test_y.drop(["index"], axis=1)

# create the model
model=Sequential()

# add layers
model.add(Dense(units=10,input_dim=train_x.shape[1],kernel_initializer='uniform',activation='relu'))
model.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))

# show the summary
model.summary()

# compile the model
model.compile(loss='binary_crossentropy',optimizer=Adam(lr=0.001),metrics=['accuracy'])

# fit the model
history=model.fit(train_x,train_y,epochs=50,batch_size=32,validation_split=0.2,verbose=1)

# accuracy
accuracy=model.evaluate(test_x, test_y)[1]*100

# print result
print("")
print("")
print("Accuracy of test:")
print(accuracy)
```
